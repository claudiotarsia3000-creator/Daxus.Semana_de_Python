```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Solicitar el código (Asegurate de escribirlo bien, ej: AAPL o PETR4.SA)
ticker = input("Escriba el ticker de la acción (ej: AAPL, TSLA, BTC-USD): ").upper()

try:
    # 2. Intentar descargar datos
    print(f"Buscando datos para {ticker}...")
    datos = yf.download(ticker, start="2024-01-01", end="2024-12-31")

    # 3. Verificación de seguridad
    if datos.empty:
        print(f"❌ ERROR: No se encontró nada para '{ticker}'.")
        print("Recomendación: Buscá el código exacto en finance.yahoo.com")
    else:
        # Extraemos solo la columna Close (Cierre)
        cotizacion_cierre = datos['Close']
        
        # 4. Graficar
        plt.figure(figsize=(10, 5))
        plt.plot(cotizacion_cierre)
        plt.title(f"Evolución de {ticker} - 2024")
        plt.grid(True)
        plt.show()
        
        print("✅ ¡Datos cargados con éxito!")

except Exception as e:
    print(f"⚠️ Ocurrió un error técnico: {e}")
```

# Hacer el analisis solicitado por el gestor
* Cotizacion Maxima
* Cotizacion Minima
* Cotizacion media de cierre de la accion 

```python
# Calcular la cotización máxima, mínima y el valor medio
maxima = round(cotizacion_cierre.max(), 2)
minima = round(cotizacion_cierre.min(), 2)
valor_medio = round(cotizacion_cierre.mean(), 2)

print(f"La cotización máxima es: {maxima}")
print(f"La cotización mínima es: {minima}")
print(f"El valor medio es: {valor_medio}")
```
