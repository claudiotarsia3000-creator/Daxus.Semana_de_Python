## LOOP/CICLOS "FOR"

```python
numeros = [1, 2, 3, 4, 5]

for numeros in numeros:
    print(numeros)
```

## LOOP / CICLO "WHILE"

```python
# Ciclo que se repite 5 veces
contador = 5

while contador > 0:
    print(contador)
    contador = contador -1
```

## Pruebas Logicas con "IF"

```python
# Un If con un Else que evalúa si el lenguaje es Python o no...
lenguaje = "Java"

if lenguaje == "Python":
    print("¡Es un gran lenguaje de programación!")
else:
    print("¡Es un lenguaje interesante!")
```

```python
# Un If que valida si el lenguaje es Python o Java
lenguaje = "Python"

if lenguaje == "Python":
    print("¡Es un gran lenguaje de programación!")
else:
    print("¡Es un lenguaje interesante!")
```
