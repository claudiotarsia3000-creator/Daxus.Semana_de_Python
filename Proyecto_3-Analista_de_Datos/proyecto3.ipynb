```python
## Ayudar a agrupar los datos por columna, en este caso por tienda
# y luego sumar el precio de cada tienda (Facturacion por tienda)

data.groupby("tienda")["precio"].sum().to_frame()
```

```python
## Ayudar a agrupar los datos por columna, en este caso por forma de pago
# y luego sumar el precio de cada forma de pago (Facturacion por forma de pago)

data.groupby("formas_de_pago")["precio"].sum().to_frame()
```

```python
## Ayudar a agrupar los datos por columna, en este caso por tienda y forma de pago
# y luego sumar el precio de cada tienda y forma de pago (Facturacion por tienda y
# forma de pago)

data.groupby("tienda")["forma_de_pago"]["precio"].sum().to_frame()
```

```python
## Guardar el resultado de la agrupacion en una variable, para luego guardarla en un archivo excel

datos_guardados = data.groupby("tienda")["forma_de_pago"]["precio"].sum().to_frame()
```

```python
## Guardar el resultado de la agrupacion en un archivo excel

datos_agrupados.to_excel("facturacion.xlsx")
```

## VISUALIZACION DE DATOS

```python
# Para graficar los datos, es necesario importar la libreria plotly
# Plotly es una libreria de graficos interactivos, que permite 
# crear graficos de barras, lineas, pastel, etc.

import plotly.express as px
```

```python
# Crear un grafico de barras, con el resultado de la agrupacion por tienda

px.histogram(data, x="tienda")
```

```python
# Crear un grafico de barras, con el resultado de la agrupacion por tienda y forma de pago

px.histogram(data,
            x="tienda",
            text_auto=True,
            title="Facturacion por tienda",
            labels={"tienda": "Tienda", "count": "Facturacion"},
            color="forma_de_pago",)

# Crear un grafico de barras, con el resultado de la agrupacion por tienda y forma de pago, pero sin color

grafico = px.histogram(data,
            x="tienda",
            text_auto=True,
            title="Facturacion por tienda",
            labels={"tienda": "Tienda", "count": "Facturacion"},
            color=None)

grafico.show()
grafico.write_html("facturacion.html")
```
