```python
import yfinance
data = yfinance.Ticker("AAPL").history("6mo")
cierre = data.Close
cierre.plot()   
```

## Hacer el analisis solicitado por el gestor
* Cotizacion Maxima
* Cotizacion Minima
* Cotizacion media de cierre de la accion 

```python
maxima = round(cierre.max(), 2)
minima = round(cierre.min(), 2)
valor_medio = round(cierre.mean(), 2)
```

```python
print(maxima)
print(minima)
print(valor_medio) 
```
