## LOOP/CICLOS "FOR"

```python
numeros = [1, 2, 3, 4, 5]

for numeros in numeros:
    print(numeros)
```

## LOOP/CICLO "WHILE"

```python
contador = 5

while contador > 0:
    print(contador)
    contador = contador -1
```

## Pruebas Logicas con "IF"

```python
# lenguaje = "Python"
lenguaje = "Java"

if lenguaje == "Python":
    print("¡Es un gran lenguaje de programación!")
else:
    print("¡Es un lenguaje interesante!")
```
