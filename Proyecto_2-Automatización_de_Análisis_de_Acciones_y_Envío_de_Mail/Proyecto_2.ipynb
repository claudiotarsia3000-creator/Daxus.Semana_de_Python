```python
import yfinance
input
ticker = input("Ingrese el código de acción: ")

input
dt_inicial = input("Introduzca la fecha de inicio (aaaa-mm-dd): ")
input
dt_final = input("Introduzca la fecha de finalización (aaaa-mm-dd): ")
datos = yfinance.Ticker(ticker)
tabla = datos.history(start=dt_inicial, end=dt_final)
tabla 
# Extraer la columna de cierre
cierre = tabla.Close
cierre.plot()
```

# Hacer el analisis solicitado por el gestor
* Cotizacion Maxima
* Cotizacion Minima
* Cotizacion media de cierre de la accion 

```python
# Calcular el valor máximo, mínimo y promedio
maxima = round(cierre.max(), 2)
minima = round(cierre.min(), 2)
valor_medio = round(cierre.mean(), 2)
# Imprimir los resultados
print(maxima)
print(minima)
print(valor_medio)
```
